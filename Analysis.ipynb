# Q1  Create a DataFrame that consists of the following columns: (1) location; (2) average deaths for2020; (3) average deaths of 2018; (4) the difference between average deaths of 2020 and 2018.The location column should include the name of the respective country

```python
locs = df.location.unique()
d2020 = []
d2018 = []
dif = []
```

```python
for loc in locs:
    d2020.append(df[df['location'] == loc]['deaths_2020_all_ages'].mean())
    d2018.append(df[df['location'] == loc]['deaths_2018_all_ages'].mean())
    dif.append(d2020[-1] - d2018[-1])
```

```python
new_df = pd.DataFrame()
```

```python
new_df['Location'] = locs
```

```python
new_df['average_deaths_2020'] = d2020
new_df['average_deaths_2018'] = d2018
new_df['average_excess_deaths'] = dif
```

```python
new_df
```

# Q2 For every country/location, find the top-5 weeks with the lowest number of excess deaths (Excess deaths are defined as the difference between the deaths that occurred in a specific week in 2020 vs 2018).


```python
df['Week_excess_deaths'] = df['deaths_2020_all_ages'] - df['deaths_2018_all_ages']
```

```python
top5s = []
for loc in locs:
    top5s.append(df[df['location'] == loc].sort_values('Week_excess_deaths').head()['Week'].values)
```

```python
top5_df = pd.DataFrame()
```

```python
top5_df['Location'] = locs
```

```python
top5s = np.array(top5s)
```

```python
top5_df['First Week'] = top5s[:,0]
top5_df['Second Week'] = top5s[:,1]
top5_df['Third Week'] = top5s[:,2]
top5_df['Fourth Week'] = top5s[:,3]
top5_df['Fifth Week'] = top5s[:,4]
```

```python
top5_df
```

```python
top5_df.to_csv('Top_5_weeks.csv')
```

# Q3 Find all the countries showing a positive excess (e.g., the diff between 2020 and 2018 greater than 0);


```python
#Excess
countries = []
for loc in locs:
    deaths_2020 = df[df['location'] == loc]['deaths_2020_all_ages'].sum()
    deaths_2018 = df[df['location'] == loc]['deaths_2018_all_ages'].sum()
    if deaths_2020 > deaths_2018:
        countries.append(loc)
```

```python
countries
```

# Q4 For each country/location, find which year had the highest mortality.


```python
years = range(2010,2022)
max_year = []
for loc in locs:
    rates = []
    for year in years:
        deaths = df[df['location'] == loc][f'deaths_{year}_all_ages'].sum()
        rates.append(deaths)
    max_year.append(years[rates.index(max(rates))])
```

```python
max_mortality_year = pd.DataFrame()
```

```python
max_mortality_year['Location'] = locs
```

```python
max_mortality_year['Max_mortality_year'] = max_year
```

```python
max_mortality_year
```

# Q5  For every week, show the country/location with the lowest mortality.

```python
years = range(2010,2022)
min_locs = []
min_deaths_list = []
for week in range(1,53):
    week_data = df[df['Week'] == week]
    locations = week_data['location'].values
    min_deaths = np.inf
    min_loc = locations[0]
    for loc in locations:
        not_na = False
        deaths = 0
        for year in years:
            d_data = week_data[week_data['location'] == loc][f'deaths_{year}_all_ages'].dropna()
            if len(d_data) > 0:
                not_na = True
                deaths += week_data[week_data['location'] == loc][f'deaths_{year}_all_ages'].sum()
        if deaths < min_deaths and not_na:
            min_deaths = deaths
            min_loc = loc
    min_locs.append(min_loc)
    min_deaths_list.append(min_deaths)    
```

```python
week_lowest = pd.DataFrame()
```

```python
week_lowest['Week'] = range(1,53)
```

```python
week_lowest['Location'] = min_locs
```
